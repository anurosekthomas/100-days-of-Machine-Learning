```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/kaggle/input/ted-talks/data.csv')
```

```python
data.head()
```

```python
#clean data
data=data.dropna()
```

```python
#change date format
data['year']=data.date.apply(lambda x:x.split(" ")[1])
data['month']=data.date.apply(lambda x:x.split(" ")[0])
monthdic={'January':'01','February':'02','March':'03','April':'04','May':'05','June':'06','July':'07','August':'08','September':'09','October':'10','November':'11','December':'12'}
data['month']=data['month'].apply(lambda x:monthdic[x])
data['date']='01'+'.'+data.month+'.'+data.year
```

```python
data.head()
```

****1.Finding most popular TEDTALK****

```python
# in terms of views
data[data.views==data.views.max()]

```

```python
#in terms of likes
data[data.likes==data.likes.max()]
```

```python
popular_shows_views=data.sort_values(by=['views'],ascending=False)
popular_shows_views
```

```python
popular_shows_likes=data.sort_values(by=['likes'],ascending=False)
popular_shows_likes
```

```python
plt.plot(data.views,data.likes)
```

**2.Finding the most popular TED talks Speaker (in terms of number of talks)**

```python
popular_speaker=data.groupby('author')['title'].count()
popular_speaker.sort_values(ascending=False)
```

****3.Month-wise Analysis of TED talk frequency****

```python
monthly_shows=data.groupby('month').size()
monthly_shows
```

```python
monthly_shows.plot(kind='bar',xlabel='months',ylabel='number of shows',title='Monthly analysis of shows')
```

**4.Year-wise Analysis of TED talk frequency**

```python
yearly_shows=data.groupby('year').size()
yearly_shows.plot(kind='bar',figsize=(12,4),color='green',xlabel='year',ylabel='number of shows',title='Yearly analysis')
```

**6.Finding TED talks with the best view to like ration**

```python
data["ratio"]=data['views']/data['likes']
data.sort_values(by='ratio',ascending=False).head()
```

**7.. Finding TED talks based on tags(like climate)**

```python
data['title']=data['title'].str.lower()
```

```python
def findtag(tag):
    shows_=[t for t in data.title if tag in t]
    print(shows_)
        
```

**8.Finding the most popular TED talks Speaker (in terms of number of views)**
